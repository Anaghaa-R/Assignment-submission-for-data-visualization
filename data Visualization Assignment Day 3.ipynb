```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from numpy.random import randn,randint,uniform,sample
%matplotlib inline
```

```python
#sns.relplot()
fmri = sns.load_dataset('fmri')
fmri.head(5)
```

```python
sns.relplot(x='timepoint',y='signal',data=fmri,color='orange',hue='region',style='region')
```

```python
#Line Plot
sns.relplot(x='timepoint',y='signal',data=fmri,color='orange',hue='region',style='region',kind='line')
```

```python
#box plot
sns.catplot(x='timepoint',y='signal',data=fmri,color='orange',hue='region',kind='box',dodge=False)
```

```python
#plotting y=x^2
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0,10)
y = x*x
plt.plot(x,y,'ro--')
plt.title("y=x^2")
plt.xlabel("x-axis",color="green")
plt.ylabel("y-axis",color='blue')
plt.show()
```

```python
#plotting sales of a shop on various days
x = [1,2,3,4,5,6,7] #days of the week 
y = [160,150,140,145,175,165,180] #sales
plt.plot(x,y,"b*--")
plt.title("Sales v/s Days of week")
plt.xlabel("Days of week",color="red")
plt.ylabel("Sales",color="purple")
plt.show()
```
